```python
# Combine relevant text columns into 'contents' and 'other_info' for each book
books['contents'] = books['description']  + ' ' + books['series'] + ' ' + books['title'] + ' ' + books['characters']
```

```python
books['other_info'] = books['genres'] + ' ' + books['publisher'] + ' ' + books['awards'] + ' ' +  books['author'] + ' ' + books['setting']
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
def stem_text(text):
    return ' '.join([ps.stem(word) for word in text.split()])
```

```python
books['contents'] = books['contents'].apply(stem_text)
```

```python
books['other_info'] = books['other_info'].apply(stem_text)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=2000,stop_words='english')
```

```python
def recommend(book_title):
    similarity = cosine_similarity(vector_other_info)
    book_index = books[books['title'] == book_title].index[0]
    distances = similarity[book_index]
    # Get top 5 similar books (excluding the book itself)
    similar_books_indices = np.argsort(distances)[::-1][1:10]
    recommended_books = books.iloc[similar_books_indices][['title', 'author', 'genres']]
    print(distances)
    return recommended_books
```

```python
def semantic_search(query, books, tfidf_matrix, tfidf_vectorizer):
    query = stem_text(query)
    query_vector = tfidf_vectorizer.transform([query])
    
    # Calculate cosine similarity between query vector and document vectors
    similarities = cosine_similarity(query_vector, tfidf_matrix)
    ranked_indices = similarities.argsort()[0][::-1]  # Sort by descending order of similarity
    print(similarities)
    # Return ranked documents based on similarity
    ranked_documents = books.iloc[ranked_indices]['title']
    print(ranked_documents)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
tfidf_vectorizer = TfidfVectorizer(max_features=5000)  # You can adjust max_features based on your dataset size
X = tfidf_vectorizer.fit_transform(books['contents'])
```

```python
silhouette_scores = []
for n_clusters in range(2, 11):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(X)
    silhouette_avg = silhouette_score(X, cluster_labels)
    silhouette_scores.append(silhouette_avg)

optimal_num_clusters = silhouette_scores.index(max(silhouette_scores)) + 2
```

```python
kmeans = KMeans(20, random_state=42)
cluster_labels = kmeans.fit_predict(X)

# Add cluster labels to the dataset
books['cluster'] = cluster_labels

# Print the number of books in each cluster
print(books['cluster'].value_counts())
```
